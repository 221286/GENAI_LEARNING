```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_community.utilities import GoogleSerperAPIWrapper
from langchain.tools import tool
from langgraph.prebuilt import create_react_agent
import os
load_dotenv()
# os.environ.get("SERPER_API_KEY")
```

```python
@tool
def callGoogleSearchApi(googleSerperAPIWrapper):
    return googleSerperAPIWrapper.run()
```

```python
llm=ChatGroq(model="llama-3.3-70b-versatile", streaming=True)
search=GoogleSerperAPIWrapper()
text="transfer your search to google"
tools=[search.run]
agent=create_react_agent(llm,tools,prompt=text)
response=agent.invoke({
    "messages": [("user","Who is the President minister of India??")]
})
print(response["messages"][-1].content)

```
