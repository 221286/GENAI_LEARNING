```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text="Islam is one of the world's major monotheistic religions, founded in the 7th century CE.It teaches belief in one God (Allah) and follows the teachings of the Prophet Muhammad.The Quran is the holy book of Islam and serves as its primary source of guidance.Muslims practice the Five Pillars, including prayer, fasting, charity, pilgrimage, and declaration of faith.Today, Islam is practiced by oer a billion people across diverse cultures and countries."
loader=RecursiveCharacterTextSplitter(chunk_size= 100, chunk_overlap= 20)
chunnks=loader.split_documents(text)
print(chunnks)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
loader=TextLoader("../data/speech.txt")
text=loader.load()
splitter=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=10)
chunks=splitter.split_documents(text)
len(chunks)
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
pdfLoader=PyPDFLoader("../data/medical_report.pdf")
pdfDoc=pdfLoader.load()
pdfSplitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=20)
pdfChunks=pdfSplitter.split_documents(pdfDoc)
len(pdfChunks)
```
