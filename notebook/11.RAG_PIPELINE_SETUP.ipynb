```python
from dotenv import load_dotenv
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_chroma import Chroma
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
load_dotenv()
```

```python
query="Machine learning and data science content"
```

```python
loader=PyPDFLoader("../data/data_science_syllabus.pdf")
text=loader.load()
splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
splittedDocs=splitter.split_documents(text)
embeddings=GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
vectorStore=Chroma.from_documents(documents=splittedDocs,embedding=embeddings,persist_directory="./vector_db/")
result=vectorStore.similarity_search(query=query)
```

```python
content=""
for doc in result:
   content+=doc.page_content+"""\n"""

content
```

```python
def getContext(query:str):
    data=vectorStore.similarity_search(query)
    content=""
    for val in data:
       content+=val.page_content +"\n"

    return {"context":content,
            "question":query}
```

llm Invoke


```python
prompt=PromptTemplate.from_template("""You're a helpful assistant provide answers based on the context for question.
                                    If you don't know the answer, then you can say I don't know.
                                    Context:{context},
                                    Question:{question}""")
```

```python
# question="I want 200 more things about machine learning"
llm=ChatGroq(model="llama-3.3-70b-versatile")
result= getContext|prompt|llm
response=llm.invoke("I want 200 more things about machine learning")
print(response.content)
```
